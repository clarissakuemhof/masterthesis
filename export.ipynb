```python
import wandb
api = wandb.Api()

# run is specified by <entity>/<project>/<run_id>
#run = api.run("clarissa-kuemhof-ludwig-maximilianuniversity-of-munich/MORL-baselines/6p56ipvd")
run = api.run("clarissa-kuemhof-ludwig-maximilianuniversity-of-munich/MORL-baselines/qextppgs")
# save the metrics for the run to a csv file
metrics_dataframe = run.history()
metrics_dataframe.to_csv("metrics.csv")
```

```python
if run.state == "finished":
    for i, row in run.history().iterrows():
      print(row["global_step"], row["eval/hypervolume"])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('metrics.csv')
#print(df)

df_cleaned = df.dropna(subset=['eval/hypervolume', 'eval/sparsity'])

# Extract relevant columns
training_steps = df_cleaned['global_step'] 
hypervolume = df_cleaned['eval/hypervolume']
sparsity = df_cleaned['eval/sparsity']

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))

# Plot Hypervolume vs Training Steps
ax1.plot(training_steps, hypervolume, marker='o', linestyle='-', color='b', label='Hypervolume')
ax1.set_xlabel('Training Steps')
ax1.set_ylabel('Hypervolume')
ax1.set_title('Hypervolume')
ax1.legend()

# Plot Sparsity vs Training Steps
ax2.plot(training_steps, sparsity, marker='o', linestyle='-', color='g', label='Sparsity')
ax2.set_xlabel('Training Steps')
ax2.set_ylabel('Sparsity')
ax2.set_title('Sparsity')
ax2.legend()

# Display the plots
plt.tight_layout()
plt.show()
```
